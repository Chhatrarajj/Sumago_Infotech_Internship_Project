```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import style
style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')

```

```python
df = pd.read_csv('Real estate.csv')
```

```python
df.info()
```

```python
df.sample(5)
```

```python
mean_price = df['Y house price of unit area'].mean()
df['PriceCategory'] = df['Y house price of unit area'].apply(lambda x: 1 if x > mean_price else 0)
```

```python
# df.drop(columns=['No', 'Y house price of unit area'], inplace=True)
```

```python
df
```

```python
x = df.drop('PriceCategory', axis=1)
y = df['PriceCategory']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train, y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5)
```
