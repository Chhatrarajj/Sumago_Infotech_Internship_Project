```python
from sklearn.preprocessing import OrdinalEncoder
Encoder = OrdinalEncoder()
df[['company','job','degree']]=Encoder.fit_transform(df[['company','job','degree']])

# x = df[['company','job','degree']]
original_data = Encoder.fit_transform(df[['company','job','degree']])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  =train_test_split(original_data,Y,test_size=0.4,random_state=5)
```
