```python
from ast import increment_lineno
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train) , (X_test , y_test) = keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train.shape
```

```python
X_train[7]
```

```python
plt.matshow(X_train[7])
```

```python
#scaling
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[5]
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened.shape
```

```python
X_train_flattened[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_pred = model.predict(X_test_flattened)
y_pred[0]
```

```python
plt.matshow(X_test[0])
```

```python
np.argmax(y_pred[0])
```

```python
# np.argmax finds max argument in array and returns index of it
y_pred_labels = [np.argmax(i) for i in y_pred]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred_labels)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#using 1 hidden layer
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_pred_hidden = model.predict(X_test_flattened)
y_pred_labels_hidden =[np.argmax(i) for i in y_pred_hidden]
cm =tf.math.confusion_matrix(labels=y_test,predictions=y_pred_labels_hidden)

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#using 2 hidden layer
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,)),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
y_pred_hidden = model.predict(X_test_flattened)
y_pred_labels_hidden =[np.argmax(i) for i in y_pred_hidden]
cm =tf.math.confusion_matrix(labels=y_test,predictions=y_pred_labels_hidden)

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
